```python
import numpy as np
import random
import os
from PIL import Image
```

```python
def split_data_in_folder(path_to_data):
    names = os.listdir(path_to_data)
    names = [os.path.join(path_to_data, f) for f in names]
    np.random.seed(615268)
    np.random.shuffle(names)
    split_1 = int(0.80 * len(names))
    split_2 = int(0.95 * len(names))
    train_data = names[:split_1]
    dev_data = names[split_1:split_2]
    test_data = names[split_2:]
    filenames = {'train': train_data,
                 'dev': dev_data,
                 'test': test_data}
    return filenames

def save_img(filenames, output_dir):
    image = Image.open(filename)
    image.save(os.path.join(output_dir, filename.split('/')[-1]))
```

```python
cracks_split = split_data_in_folder('road_images/cracks')
```

```python
#Create necessary directory for train/dev/test, save images there

for sp in ['train', 'dev', 'test']:
    output_dir_split = os.path.join('road_images/split', sp)   
    if not os.path.exists(output_dir_split):
        os.mkdir(output_dir_split)
    else:
        print("Warning: dir {} already exists".format(output_dir_split))
        for filename in cracks_split[sp]:
            save_img(filename, output_dir_split)
```
