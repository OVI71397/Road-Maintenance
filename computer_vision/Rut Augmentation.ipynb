## Augment

```python
# Horizontal flip
img = cv2.imread('road_images/rut/1.jpeg', cv2.IMREAD_COLOR)
horizont_flip = cv2.flip(img, 1)
plt.figure(figsize=[18,5])
plt.subplot(141);plt.imshow(img[:,:,::-1]);plt.title("Original");
plt.subplot(142);plt.imshow(horizont_flip[:,:,::-1]);plt.title("Vertical Flip");
```

```python
default_path = 'road_images/rut/'
for image in glob.glob("road_images/rut/*.jpeg"):
    instance = cv2.imread(image, cv2.IMREAD_COLOR)
    hor_flip = cv2.flip(instance, 1)
    new_name = image.split('/')[-1].split('.')[0] + '_hor_flip.jpeg'
    cv2.imwrite(os.path.join(default_path, new_name), hor_flip)
```

```python
img_1 = cv2.imread('road_images/rut/000003.jpeg', cv2.IMREAD_COLOR)
img_2 = cv2.imread('road_images/rut/000025.jpeg', cv2.IMREAD_COLOR)
img_3 = cv2.imread('road_images/rut/road rut_9.jpeg', cv2.IMREAD_COLOR)
matrix1 = np.ones(img_1.shape, dtype="uint8")*50
matrix2 = np.ones(img_2.shape, dtype="uint8")*50
matrix3 = np.ones(img_3.shape, dtype="uint8")*50

img_1_brighter = cv2.add(img_1, matrix1)
img_2_brighter = cv2.add(img_2, matrix2)
img_3_brighter = cv2.add(img_3, matrix3)

fig = plt.figure(figsize=[18,5])
fig.add_subplot(2, 3, 1);plt.imshow(img_1[:,:,::-1]);
fig.add_subplot(2, 3, 2);plt.imshow(img_2[:,:,::-1]);
fig.add_subplot(2, 3, 3);plt.imshow(img_3[:,:,::-1]);
fig.add_subplot(2, 3, 4);plt.imshow(img_1_brighter[:,:,::-1]);
fig.add_subplot(2, 3, 5);plt.imshow(img_2_brighter[:,:,::-1]);
fig.add_subplot(2, 3, 6);plt.imshow(img_3_brighter[:,:,::-1]);
```

```python
# Brightness adjustment
for image in glob.glob("road_images/rut/*.jpeg"):
    instance = cv2.imread(image, cv2.IMREAD_COLOR)
    matrix = np.ones(instance.shape, dtype="uint8")*50
    instance_brighter = cv2.add(instance, matrix)
    new_name = image.split('/')[-1].split('.')[0] + '_brighter.jpeg'
    cv2.imwrite(os.path.join(default_path, new_name), instance_brighter)
```
